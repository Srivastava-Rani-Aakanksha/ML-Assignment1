# # Function for calculating Mean Squared Error
 

```python
def err(coeff,x_test,y_test):
    predi=np.dot(x_test,coeff)
    mse=np.mean(np.square(predi-y_test))
    print("Mean Square Error: "+str(mse))
```

# 4.Multivariate Linear Regression : Gradient Descent

```python
y_train.shape
```

```python
def LinearRegressionMultivariate(X,Y,learning_rate,iteration):
    m=Y.size #number of rows(test entries)
    n=X.shape[1] #number of features
    Y=Y.reshape((m,1))
    M_wg=np.zeros((n,1)) #column vector of gradient descent parameters
    
    cost_h=[]
    
    for i in range(iteration):
        y_pred_MG=np.dot(X,M_wg)
        cost=(1/(2*m))*np.sum(np.square(y_pred_MG - Y))
        
        partial_diff=(1/m)*np.dot(X.T,y_pred_MG - Y)
        
        M_wg=M_wg-learning_rate*partial_diff
        
        cost_h.append(cost)
        
        if i % (iteration // 10)==0:
            print("Cost=",cost)
    return M_wg,cost_h
```

```python
alpha=0.05
iteration=10000
M_wg,cost=LinearRegressionMultivariate(x_train,y_train,alpha,iteration)
```
